## ***openfold***

```python
import subprocess
import os

# Directory where the repo is to be cloned
repo_dir = "openfold"

# Check if the directory already exists
if os.path.isdir(repo_dir):
    print(f"The directory '{repo_dir}' already exists. Removing it to proceed.")
    subprocess.run(['rm', '-rf', repo_dir], check=True)

# Define the commands to run
commands = f"""
git clone https://github.com/aqlaboratory/openfold.git &&
cd openfold &&
sed -i 's/std=c++14/std=c++17/g' setup.py &&
conda run -n prolig_0001 pip install . --no-deps --no-cache-dir
"""

# Execute the commands
try:
    result = subprocess.run(commands, shell=True, check=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True, executable="/bin/bash")
    print("Commands executed successfully:")
    print(result.stdout)
except subprocess.CalledProcessError as e:
    print("An error occurred:")
    print(e.stderr)
```

# **installation other dependencies**

```python
import subprocess

# Define the path to the environment YAML file
env_yaml_path = "/home/jupyter/google_cloud/protein_ligand/neuralPlexer/NeuralPLexer/NeuralPLexer.yml"

# Define the conda environment name
env_name = "prolig_0001"

# Prepare the commands
update_command = f"conda env update -n {env_name} --file {env_yaml_path}"
clean_command = "conda clean -afy"

# Execute the update command
try:
    update_result = subprocess.run(update_command, shell=True, check=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True, executable="/bin/bash")
    print("Environment update executed successfully:")
    print(update_result.stdout)
except subprocess.CalledProcessError as e:
    print("An error occurred during environment update:")
    print(e.stderr)
    exit(1)  # Exit if the update fails to prevent attempting cleanup on a failed update

# Execute the clean command
try:
    clean_result = subprocess.run(clean_command, shell=True, check=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True, executable="/bin/bash")
    print("Conda clean executed successfully:")
    print(clean_result.stdout)
except subprocess.CalledProcessError as e:
    print("An error occurred during conda clean:")
    print(e.stderr)
```

## **final installation of neuralplexer**

```python
import subprocess

# Define the conda environment name
env_name = "prolig_0001"

# Define the directory path where the make command should be run
dir_path = "/home/jupyter/google_cloud/protein_ligand/neuralPlexer/NeuralPLexer"

# Commands to run
# Includes changing to the target directory before executing 'make install'
commands = f"cd {dir_path} && conda run -n {env_name} make install"

# Execute the commands
try:
    # Use subprocess.run() to execute the commands
    result = subprocess.run(commands, shell=True, check=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True, executable="/bin/bash")
    print("Commands executed successfully:")
    print(result.stdout)
except subprocess.CalledProcessError as e:
    print("An error occurred:")
    print(e.stderr)
```

# **download model**

```python
import subprocess

# URL of the file to download
url = "https://zenodo.org/records/10373581/files/neuralplexermodels_downstream_datasets_predictions.zip?download=1"

# Target directory where the file will be saved
target_directory = "/home/jupyter/google_cloud/protein_ligand/neuralPlexer"

# Command to download the file using wget
# -P specifies the directory where the file will be saved
command = ["wget", "-P", target_directory, url]

# Execute the command
result = subprocess.run(command, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)

# Check the result
if result.returncode == 0:
    print("Download successful.")
else:
    print("Download failed.")
    # Print error messages if any
    print("Error message:", result.stderr)
```
